# Detail data pengguna TransJakarta
Data ini berisikan informasi mengenai transaksi penggunaan Transjakarta yang diantara lain:
1. transID: ID transaksi unik untuk setiap transaksi.
2. payCardID: Identifikasi utama pelanggan. Kartu yang digunakan pelanggan sebagai tiket untuk masuk dan keluar.
3. payCardBank: Nama penerbit bank dari kartu pelanggan.
4. payCardName: Nama pelanggan yang tertanam di kartu.
5. payCardSex: Jenis kelamin pelanggan yang tertanam di kartu.
6. payCardBirthDate: Tahun kelahiran pelanggan.
7. corridorID: ID Koridor / ID Rute sebagai kunci untuk pengelompokan rute.
8. corridorName: Nama Koridor / Nama Rute yang mencakup titik awal dan akhir untuk setiap rute.
9. direction: 0 untuk pergi, 1 untuk kembali. Arah dari rute.
10. tapInStops: ID Pemberhentian Tap In (masuk) untuk mengidentifikasi nama pemberhentian.
11. tapInStopsName: Nama Pemberhentian Tap In (masuk) tempat pelanggan melakukan tap in.
12. tapInStopsLat: Lintang dari Pemberhentian Tap In.
13. tapInStopsLon: Bujur dari Pemberhentian Tap In.
14. stopStartSeq: Urutan pemberhentian, pemberhentian pertama, pemberhentian kedua, dll. Terkait dengan arah.
15. tapInTime: Waktu tap in. Tanggal dan waktu.
16. tapOutStops: ID Pemberhentian Tap Out (keluar) untuk mengidentifikasi nama pemberhentian.
17. tapOutStopsName: Nama Pemberhentian Tap Out (keluar) tempat pelanggan melakukan tap out.
18. tapOutStopsLat: Lintang dari Pemberhentian Tap Out.
19. tapOutStopsLon: Bujur dari Pemberhentian Tap Out.
20. stopEndSeq: Urutan pemberhentian, pemberhentian pertama, pemberhentian kedua, dll. Terkait dengan arah.
21. tapOutTime: Waktu tap out. Tanggal dan waktu.
22. payAmount: Jumlah yang dibayar pelanggan. Ada yang gratis, ada yang tidak.

# Cleaned Data

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Tap In Time Distribution
Visualisasi jam berapa pelanggan sering menaiki Transjakarta

```python
# Plot distribusi jam tap in
plt.figure(figsize=(12, 6))
sns.histplot(df['tapInHour'], bins=24, kde=True)
plt.title('Distribusi Jam Tap In')
plt.xlabel('Jam')
plt.ylabel('Frekuensi')
plt.xticks(range(24))
plt.show()
```

# Comparison of Tap In and Tap Out by Location
Perbandingan tapIn dan tapOut berdasarkan lokasi


```python

# Menggabungkan data tap in dan tap out
tap_in_out = pd.DataFrame({
    'Tap In': df['tapInStopsName'].value_counts(),
    'Tap Out': df['tapOutStopsName'].value_counts()
}).fillna(0)

# Tampilkan tabel
tap_in_out
```

# Average Number of Payments by Gender
Rata rata pengguna Transjakarta berdasarkan jenis kelamin

```python
average_payment_by_gender = df.groupby('payCardSex')['payAmount'].mean().reset_index()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='payCardSex', y='payAmount', data=average_payment_by_gender)
plt.title('Rata-rata Jumlah Pembayaran Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Rata-rata Pembayaran')
plt.show()
```

# Transaction Frequency by Age
Frekuensi usia pengguna Transjakarta


```python
# Adding age column base on their birthdate
df['age'] = 2024 - df['payCardBirthDate']

# Make category of age
bins = [0, 17, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '66+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Counting age
age_group_counts = df['age_group'].value_counts().sort_index()

# Creating pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Frekuensi Transaksi Berdasarkan Usia')
plt.show()
```
